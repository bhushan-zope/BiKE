```python
import pandas as pd
import glob
import os

path = 'path-to-data-repository'
```

```python
df = pd.read_csv('{}query03-05.csv'.format(path))
df = df.fillna('')
df = df.astype({
    'doi': 'str', 'bioActivity': 'str', 'molType': 'str', 'collectionSpecie': 'str', 'collectionSite': 'str',
    'collectionType': 'str', 'molecularMass': 'str', 'monoisotropicMass': 'str', 'cLogP': 'str', 'tpsa': 'str',
    'numberOfLipinskiViolations': 'str', 'numberOfH_bondAcceptors': 'str', 'numberOfH_bondDonors': 'str',
    'numberOfRotableBonds': 'str', 'molecularVolume': 'str',
})
df = df[(df['molType'].str.contains('http://nubbe.db/')) | (df['molType'] == '')]
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
smiles_names = pd.read_csv('{}smiles_names03-05.csv'.format(path))
smiles_names
```

```python
df = df.join(smiles_names.set_index('smiles'), on='smile')
df = df.fillna('')
df
```

```python
all_files = glob.glob(os.path.join("{}pdfs/".format(path), "*.pdf"))
all_files_df = pd.DataFrame([file.replace("{}pdfs/".format(path), '') for file in all_files], columns=['file_name'])
all_files_df['doi'] = all_files_df['file_name'].str.replace('@', '/')
all_files_df['doi'] = all_files_df['doi'].str.replace('.pdf','')
dowloaded_df = df.join(all_files_df.set_index('doi'), on='doi', how='inner').reset_index(drop=True)
dowloaded_df.to_parquet('{}file-name_query03-05.parquet'.format(path))
dowloaded_df
```

```python
# not downloaded but on query
pd.DataFrame(df['doi'][~df['doi'].isin(all_files_df['doi'])].unique(), columns=['not_downloaded'])
```

```python
# not on query but downloaded
pd.DataFrame(all_files_df['doi'][~all_files_df['doi'].isin(df['doi'])].to_list(), columns=['downloaded'])
```

```python
import fitz

texts = []
for file_name in dowloaded_df['file_name'].unique():
    doc = fitz.open(f"{path}pdfs/{file_name}")
    texts.append('')
    for page_number in range(doc.page_count):
        page = doc.load_page(page_number)
        texts[-1] = texts[-1] + page.get_text("text")
texts_df = pd.DataFrame({'doi': dowloaded_df['doi'].unique(), 'file_name': dowloaded_df['file_name'].unique(), 'text': texts})
texts_df
```
